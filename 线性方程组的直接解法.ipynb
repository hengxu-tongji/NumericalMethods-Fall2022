# 1.两点边值问题离散方程的解

### （1）编写利用克洛托分解求解上述线性方程组的函数 mycrout，其中 0 = 200

```python
import numpy as np
import time
from scipy.linalg import solve

def mycrout(A,b):
    start=time.time()
    n = A.shape[0]
    L = np.zeros((n,n))
    U = np.zeros((n,n))
    for i in range(n):
        U[i,i] = 1
    for i in range(n):
        for j in range(i,n):
            L[j,i] = A[j,i] - np.dot(L[j,:i],U[:i,i])
        for j in range(i+1,n):
            U[i,j] = (A[i,j] - np.dot(L[i,:i],U[:i,j]))/L[i,i]
    y = np.empty_like(b)
    for i in range(n):
        y[i]=(b[i]-L[i,0:i].dot(y[0:i]))/L[i,i]
    x = np.empty_like(b)
    for i in range(n-1,-1,-1):
        x[i]=(y[i]-U[i,i+1:n].dot(x[i+1:n]))/U[i,i]
    end=time.time()
    total=end-start
    print(total)
    return x

n = 200
A = np.diag(4*np.ones(n))+np.diag((-1)*np.ones(n-1),-1)+np.diag((-1)*np.ones(n-1),1)
b = np.ones(n)*200
b[0],b[n-1]=100,100
x=mycrout(A,b)
print(x)
```

### (2) 编写利用平方根法求解上述线性方程组的函数 mysqroot，其中 n = 200;

```python
import numpy as np
import time

def mychase(a,b,c,d):
    start=time.time()
    n=d.size
    l=np.zeros(n)
    u=np.zeros(n-1)
    y=np.zeros(n)
    x=np.zeros(n)
    l[0]=b[0]
    y[0]=d[0]/l[0]
    u[0]=c[0]/l[0]
    for i in range(1,n-1):
        l[i]=b[i]-a[i-1]*u[i-1]
        y[i]=(d[i]-y[i-1]*a[i-1])/l[i]
        u[i]=c[i]/l[i]
    l[n-1] = b[n-1]-a[n-2]*u[n-2]
    y[n-1] = (d[n-1]-y[n-2]*a[n-2])/l[n-1]
    
    x[n-1] = y[n-1]
    for i in range(n-2,-1,-1):
        x[i]=y[i]-u[i]*x[i+1]
    end=time.time()
    total=end-start
    print(total)
    return x

n = 200
b = 4*np.ones(n)
a = (-1)*np.ones(n-1)              
c = (-1)*np.ones(n-1)
d = np.ones(n)*200
d[0],d[n-1]=100,100
x=mychase(a,b,c,d)
print(x)
```

### （3） 编写追赶法求解上述线性方程组的函数 mychase，其中 n = 200;

```python
import numpy as np
import time
from scipy.linalg import solve

def mysqroot(A,b):
    start=time.time()
    A1=A.copy()
    n = np.shape(A1)[0]
    G = np.zeros((n,n))
    for i in range(n):
        G[i,i]=np.sqrt(A1[i,i])
        u_t=(1/G[i,i])*A1[i,i+1:]
        U=np.outer(u_t,u_t)
        G[i,i+1:]=u_t
        A1[i+1:,i+1:]=A1[i+1:,i+1:]-U 
    L=G.T
    U=G
    y = np.empty_like(b)
    for i in range(n):
        y[i]=(b[i]-L[i,0:i].dot(y[0:i]))/L[i,i]
    x = np.empty_like(b)
    for i in range(n-1,-1,-1):
        x[i]=(y[i]-U[i,i+1:n].dot(x[i+1:n]))/U[i,i]
    end=time.time()
    total=end-start
    print(total)
    return x

n = 200
A = np.diag(4*np.ones(n))+np.diag((-1)*np.ones(n-1),-1)+np.diag((-1)*np.ones(n-1),1)
b = np.ones(n)*200
b[0],b[n-1]=100,100
x=mysqroot(A,b)
print(x)
```
